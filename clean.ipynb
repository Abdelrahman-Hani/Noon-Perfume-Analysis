```python
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df['num_seller_ratings'] = (
    df['num_seller_ratings']
      .str.replace('K', '', regex=False)
      .astype(float)
      * 1000
      ).astype(int)
```

```python
df['discount_amount'] = df['old_price'] - df['new_price']
```

```python
df['discount_percentage'] = (
    (df['old_price'] - df['new_price']) / df['old_price'] * 100
).round(2)
```
